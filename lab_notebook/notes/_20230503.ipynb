## Goal of this analysis

To create a Seurat object of all the single-cell MB tumour samples from Vladoiu et al. 2019.


## Files and directories

### Code

See `<tumour Vladoiu software directory>/create_srat.R`.

### Output

The following bash command was run to generate the Seurat object (the working directory was set as the output directory).


```{bash}
$ Rscript /CBL_scRNAseq/software/tumour/Vladoiu/create_srat.R \
      --in_dir /isilon/CBL_scRNAseq-archived/data/tumour/Vladoiu/ \
      --out_dir ./ \
      &> create_srat.log
```


The figures are screenshots of the output after running 


## Results
